1. FizzBuzz

```python
for i in range(1, 100):
  if i%3==0:
    print('Fizz')
  elif i%5==0:
    print('Buzz')
  elif i%3==0 and i%5==0:
    print('FizzBuzz')
  else:
    print(i)
```

2. Electricity Bill calculator

```python
u = int(input('Enter the number of units consumed :'))
charge = 0
if u <= 100:
  charge = u*5
elif u <= 200:
  charge = (100*5) + ((u-100)*7)
elif u <= 300:
  charge = (100*5) + (100*7) + ((u-200)*10)
  print('Total Bill Amount :',charge)
```

3. Display a pattern for a simple progress indicator


```python
char = "*"
for i in range(1,6):
  print(char*i)
```

4. Palindrome

```python
n = input('Enter a number :')
if n == n[::-1]:
  print('The given number is a palindrome')
else:
  print('The given number is not a palindrome')
```
